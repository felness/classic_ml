```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

def f(x):
    return x**4 - 10 * x + 3
```

```python
def grad_f(x):
    return 4* (x ** 3) - 10
```

```python
x_values = [x for x in np.arange(-10,10, 0.2)]
y_values = [f(x) for x in x_values]
```

```python
plt.figure(figsize=(12,8))
plt.plot(x_values, y_values)
plt.xlabel('min_x')
plt.ylabel('f(x)')
```

```python
def gradient(x0, epsilon, num_iterations, learning_rate):
    x_start = x0
    for x in range(num_iterations):
        x_new = x_start - learning_rate* grad_f(x_start)
        if abs(x_new - x_start) < epsilon:
            return x_start
        x_start = x_new
    return x_start
```

```python
eps = 0.001
l_r = 0.01
x0 = 5
n = 10000
x_min = gradient(x0, eps, n, l_r)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x_values, y_values)
plt.xlabel('min_x')
plt.ylabel('f(x)')
plt.scatter(x_min,f(x_min), marker='o', c='red')
```
